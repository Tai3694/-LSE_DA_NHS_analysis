```python
# Import Pandas and NumPy.
import pandas as pd
import numpy as np
```

```python
# Read data files from the current working directory.
ad = pd.read_csv('actual_duration.csv')

ar = pd.read_csv('appointments_regional.csv')

nc = pd.read_excel('national_categories.xlsx')
```

**Question 1:** Between what dates were appointments scheduled? 

```python
# View the first five rows of appointment_date for the ad DataFrame
ad.head(5)
```

```python
# View the first five rows of appointment_date for the nc DataFrame
nc.head(5)
```

```python
# Change the date format of ad['appointment_date'].
ad['appointment_date'] = pd.to_datetime(ad['appointment_date'])

# View the DataFrame.
ad.info()

#View the DataFrame
ad.head(5)
```

```python
# Change the date format of ad['appointment_date'].
ar['appointment_month'] = pd.to_datetime(ar['appointment_month'])

# View the DataFrame.
ar.info()

#View the DataFrame
ar.head(5)
```

```python
# Determine the minimum and maximum dates in the ad DataFrame.
print (ad.appointment_date.min())
print (ad.appointment_date.max())

# Determine the minimum and maximum dates in the nc DataFrame.
print (nc.appointment_month.min())
print (nc.appointment_month.max())
```

**Question 2:** Which service setting was the most popular for NHS North West London from 1 January to 1 June 2022?


```python
#Table of the data 1st Jan 2022 to 1st June 2022 for North West London
nc2 = nc.loc[((nc['appointment_date'] >= '2022-1-1') & (nc['appointment_date'] <= '2022-6-1') & (nc['icb_ons_code'] == 'E54000027'))]
print(nc2)
```

```python
#Highest count of each service setting
nc2.groupby('service_setting').count_of_appointments.nlargest(10).reset_index()
```

**Question 3:** Which month had the highest number of appointments?

```python
#Method 1
nc.groupby(pd.Grouper(key='appointment_date',freq='M')).agg({'count_of_appointments':'sum'})
```

```python
#Method 2
nc.groupby(nc['appointment_date'].dt.strftime('%B'))['count_of_appointments'].sum().sort_values()
```

**Question 4:** What was the total number of records per month?

```python
nc['appointment_date'].groupby(nc.appointment_date.dt.to_period("M")).agg('count')
```

```python
#Create new DataFrame
tweets_text = tweets[['tweet_full_text']]

#View DataFrame
print(tweets_text)
tweets_text.head(5)
```

```python
#Create new variable 
tags = []
```

```python
#Loop
for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
#Create Pandas Series and use value Counts
myseries = pd.Series(tags)
results = myseries.value_counts()

print(results)
```

```python
#Display 30 records 
results.head(30)
```

**Identify top trending hastags**

```python
#Create new DataFrame
data = pd.DataFrame({'Word':results.index, 'Count':results.values})
print(data)
```

```python
#Look at count > 10
data2 = data[data.Count > 10]
print(data2)
```

```python
#Create barplot - Tag Count >10 
figure(num=None, figsize=(20,18), dpi=80, facecolor='w', edgecolor='r')
sns.barplot(x='Count', y='Word', data=data2)
plt.show()

#Save plot
plt.savefig('Top 10 Tags.png')
```

```python
#Dataframe looking at top 10
data3 = data2.head(10)
print(data3)
```

```python
#Creat barplot - Frequent tags  
figure(num=None, figsize=(15,12), dpi=80, facecolor='w', edgecolor='r')
sns.barplot(x='Word', y='Count', data=data3)

#Save plot
plt.savefig('Frequent Tags.png')
```

**Question 1: Should the NHS start looking at increasing staff levels?**

```python
#Using groupby()
ar_agg = ar.groupby(['appointment_month',
                           'hcp_type',
                          'appointment_status',
                          'appointment_mode',
                          'time_between_book_and_appointment']).sum([]).reset_index()

#View DataFrame
print(ar_agg)
```

```python
#Create new DataFrame
ar_df = ar2.groupby(['appointment_month']).sum(['count_of_appointments'])

#View dataframe
ar_df

#New column 
ar_df['utilisation'] = ar_df['count_of_appointments']/30

#View dataframe
print(ar_df)
```

```python
#Round utiltisation to one decimal place 
ar_df['utilisation2'] = ar_df['utilisation'].round(decimals = 1)

print(ar_df)
```

```python
#Set the colour/plot style
plt.figure(figsize=(15, 12))
sns.set_style('white')

#Create lineplot - monthly visits
sns.lineplot(x='appointment_month', y='count_of_appointments',
             data=ar_df, ci=None)

#Save plot
plt.savefig('Number of monthly visits.png')
```

```python
#Set the colour/plot style
plt.figure(figsize=(15, 12))
sns.set_style('white')

#Create line plot - capacity utilisation 
sns.lineplot(x='appointment_month', y='utilisation2',
             data=ar_df, ci=None)

#Save plot
plt.savefig('Average utilisation of services.png')
```
